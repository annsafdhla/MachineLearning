```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.scatter(siswa['Rata-rata nilai IPA'],siswa['Rata-rata nilai IPS'])
plt.xlabel("Rata-rata nilai IPA")
plt.ylabel("Rata-rata nilai IPS")
plt.title("Grafik Nilai Siswa")
```

Data Training

```python
x_train = siswa[['Rata-rata nilai IPA','Rata-rata nilai IPS']].values
```

Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

Clustering

```python
from sklearn.cluster import KMeans
kmean = KMeans (n_clusters=2)
kmean
```

```python
y_cluster = kmean.fit_predict(x_train)
```

```python
y_cluster
```

```python
siswa['Cluster'] = y_cluster
```

```python
siswa
```

```python
kmean.cluster_centers_
```

Visualisasi Hasil Data Clustering

```python
plt.figure(figsize=(10,5))
plt.scatter(x_train[:,0],x_train[:,1],c=kmean.labels_)
plt.scatter(kmean.cluster_centers_[0,0],kmean.cluster_centers_[0,1],cmap='rainbow',marker='*',s=150,label='Centroid 0')
plt.scatter(kmean.cluster_centers_[1,0],kmean.cluster_centers_[1,1],cmap='rainbow',marker='*',s=150,label='Centroid 1')
plt.xlabel("Rata-rata Nilai IPA")
plt.ylabel("Rata-rata Nilai IPS")
plt.legend()
```

Mencari K (Jumlah Cluster) Terbaik

```python
inertias = []
k_range = range(1,10)
for k in k_range:
  km = KMeans(n_clusters=k).fit(x_train)
  inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.xlabel("K")
plt.ylabel("Sum Of Error")
plt.plot(k_range,inertias)
plt.grid()
```
